```python
import ast

def covert(obj):
    l = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            l.append(i["name"])
            counter += 1
        else:
            break
    return l
```

```python
import ast

def fetch_director(obj):
    l = []
    counter = 0
    for i in ast.literal_eval(obj):
        if i["job"]=="Director":
            
            l.append(i["name"])
            break
    return l
```

```python
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()  # ✅
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
def recommend(movie):
    # Convert the movie column to lowercase and the input to lowercase
    movie_lower = movie.lower()
    titles_lower = new["title"].str.lower()
    
    # Check if the movie exists
    if movie_lower not in titles_lower.values:
        print(f"Movie '{movie}' not found in dataset.")
        return
    
    # Get the index of the movie
    movie_index = new[titles_lower == movie_lower].index[0]
    
    # Compute distances and get top 5 recommendations
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), key=lambda x: x[1], reverse=True)[1:6]
    
    # Print recommended movies
    for i in movies_list:
        print(new.iloc[i[0]].title)
```

```python
recommend("Batman begins")
```
